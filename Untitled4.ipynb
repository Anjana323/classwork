```python
import plotly.express as px
import pandas as pd
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv("wine.data.csv")

# Assuming the first column is the target (class), and the rest are features
target_col = df.columns[0]  # First column as target
features = df.columns[1:]    # All other columns as features

# Perform PCA
pca = PCA()
components = pca.fit_transform(df[features])

# Create DataFrame for transformed components
components_df = pd.DataFrame(components, columns=[f"PC{i+1}" for i in range(len(features))])
components_df["class"] = df[target_col]  # Add class labels

# Label principal components with explained variance
labels = {
    f"PC{i+1}": f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

# Create scatter matrix
fig = px.scatter_matrix(
    components_df,
    dimensions=[f"PC{i+1}" for i in range(4)],  # First 4 PCs for visualization
    color="class",
    labels=labels
)
fig.update_traces(diagonal_visible=False)
fig.show()

```
